```python
import requests
```

```python
import pandas as pd, numpy as np
```

```python
data=pd.read_csv("movie.csv")
```

```python
ids=data.movie_id.tolist()
```

```python
def posters(n1,n2):
  list_pstrs=[]
  for movie_id in ids[n1:n2]:
    url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key=f8ecd9361b4bb5f50fedd44ab74d585a"
    response = requests.get(url)
    poster_paths = response.json().get("poster_path")
    list_pstrs.append([poster_paths,movie_id])
  return list_pstrs
```

```python
posters_list_500=posters(0,500)
```

```python
posters_list_1000=posters(500,1000)
```

```python
posters_list_1500=posters(1000,1500)
```

```python
posters_list_2000=posters(1500,2000)
```

```python
posters_list_2500=posters(2000,2500)
```

```python
posters_list_3000=posters(2500,3000)
```

```python
posters_list_3500=posters(3000,3500)
```

```python
posters_list_4000=posters(3500,4000)
```

```python
posters_list_4500=posters(4000,4500)
```

```python
posters_list_4803=posters(4500,4803)
```

```python
posters_list_complete=posters_list_500+posters_list_1000+posters_list_1500+posters_list_2000+posters_list_2500+posters_list_3000+posters_list_3500+posters_list_4000+posters_list_4500+posters_list_4803
```

```python
id_poster=[]
poster_path_poster=[]
for i in posters_list_complete:
  id_poster.append(i[1])
  poster_path_poster.append(i[0])
```

```python
postdt=pd.DataFrame({'id':id_poster,'poster_path':poster_path_poster})
```
